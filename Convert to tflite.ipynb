```python
import os
from google.colab import files

# Cari file .tflite di dalam folder runs/detect
tflite_path = "my_model_saved_model/my_model_float32.tflite"
for root, dirs, filenames in os.walk("runs/detect"):
    for file in filenames:
        if file.endswith(".tflite"):
            tflite_path = os.path.join(root, file)
            break

# Tampilkan path dan download file
if tflite_path:
    print("‚úÖ File ditemukan:", tflite_path)
    files.download(tflite_path)
else:
    print("‚ùå model.tflite tidak ditemukan. Pastikan konversi berhasil.")
```
